```python
import pandas as pd
```

### Perfil eleitor

```python
df_perfil_eleitor = pd.read_csv('perfil_eleitor_secao_2020_RS.csv', delimiter=';', encoding='latin1')

for column in df_perfil_eleitor.columns.values:
    print(column)
```

```python
df = df_perfil_eleitor.copy()

df['DS_FAIXA_ETARIA'] = df['DS_FAIXA_ETARIA'].str.strip()

df = df.groupby(
    ['NM_MUNICIPIO', 'NR_ZONA', 'DS_GENERO', 'DS_ESTADO_CIVIL', 'DS_FAIXA_ETARIA', 'DS_GRAU_ESCOLARIDADE'], 
    as_index=False
)['QT_ELEITORES_PERFIL'].sum()

df['ID_MUNZONA'] = df['NM_MUNICIPIO'] + '_' + df['NR_ZONA'].astype(str)

df = df[[
    'ID_MUNZONA',
    'NM_MUNICIPIO',
    'NR_ZONA',
    'DS_GENERO',
    'DS_ESTADO_CIVIL',
    'DS_FAIXA_ETARIA',
    'DS_GRAU_ESCOLARIDADE',
    'QT_ELEITORES_PERFIL'
    ]]

df.head()
```
