```python
import pandas as pd
```

```python
df_dimpz = pd.read_csv('saidas/perfil_eleitor_tratado.csv', delimiter=',')

for column in df_dimpz.columns.values:
    print(column)
```

```python
print(df_dimpz['ID_MUNZONA'].value_counts())
print()
df_dimpz.head()
```

```python
df_dimpz[df_dimpz['ID_MUNZONA'] == 'ACEGUÁ_7'].sort_values(by=['QT_ELEITORES_PERFIL'], ascending=False)
```

```python
def moda(df, att, peso='QT_ELEITORES_PERFIL'):
    return df.groupby(att)[peso].sum().idxmax()

df_moda = df_dimpz.groupby(['ID_MUNZONA', 'NM_MUNICIPIO', 'NR_ZONA']).apply(lambda df_dimpz: pd.Series({
    'DS_GENERO': moda(df_dimpz, 'DS_GENERO'),
    'DS_ESTADO_CIVIL': moda(df_dimpz, 'DS_ESTADO_CIVIL'),
    'DS_FAIXA_ETARIA': moda(df_dimpz, 'DS_FAIXA_ETARIA'),
    'DS_GRAU_ESCOLARIDADE': moda(df_dimpz, 'DS_GRAU_ESCOLARIDADE')
})).reset_index()

df_moda.head()
```
